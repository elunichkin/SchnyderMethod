```python
import networkx as nx
import matplotlib.pyplot as plt
from schnyder import draw
%matplotlib inline

"""
Write in file two numbers V and E: number of vertices and edges in G, then edges (u, v) of G, 2 numbers at a line:

Example graph:
6 9
0 1
1 2
2 3
3 4
4 0
0 5
1 5
2 5
3 5
4 5
"""

try:
    G, realizer = draw('input.txt')
    for node, coord in realizer.items():
        G.node[node]['pos'] = coord
    pos = nx.get_node_attributes(G, 'pos')
    print(pos)
    nx.draw_networkx(G, pos)
except:
    print('Error: Graph can\'t be drawn')
```
